Base response to **"What is a good check for rest/talk raikou?"**

Rest/Talk Raikou is a set that utilizes the combination of Rest and Sleep Talk to provide reliable recovery and the ability to attack even while asleep. To counter or check this Raikou set effectively, you'll want to consider the following options:

- Toxic Users: **WRONG**

Pokémon that can use Toxic are effective against Rest/Talk Raikou. Poisoning Raikou will bypass its ability to heal off status conditions with Rest.
Taunt Users:

Pokémon with access to Taunt can prevent Rest/Talk Raikou from using Sleep Talk and healing with Rest. This limits its recovery options and makes it vulnerable to continuous attacks.

- Status Inflictors: **WRONG**

Pokémon that can inflict status conditions like paralysis or burn can hinder Rest/Talk Raikou's effectiveness. A burned Raikou, for example, will have its Attack stat halved **WRONG! Raikou is a special attacker**, reducing the damage it deals.

- Ground-types: 

Ground-type Pokémon are immune to Raikou's Electric-type STAB moves, and they are often bulky enough to take its attacks. They can then proceed to threaten Raikou with super effective Ground-type moves.

- Special Walls:

Pokémon with high Special Defense can often handle Raikou's attacks well, even if it manages to get a boost from Calm Mind. Examples include Blissey and Snorlax.


- Phazing Moves:

Pokémon with phazing moves like Roar or Whirlwind can force Raikou to switch out, disrupting its Rest and Sleep Talk cycle. Skarmory is a notable example with access to Whirlwind.

- Pressure Pokémon:

Pokémon with the ability Pressure can increase the PP consumption of Raikou's moves, potentially causing it to run out of viable options more quickly.

- Critical Hits:

Moves that increase the likelihood of scoring critical hits, such as Focus Energy or Dire Hit, can help bypass Raikou's increased Defense from Calm Mind.

```python
from langchain_community.document_loaders import DirectoryLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain_community.vectorstores import Chroma

from langchain.embeddings.openai import OpenAIEmbeddings

import os
from dotenv import load_dotenv
load_dotenv()
```

## load and chunk text


```python
loader = DirectoryLoader(
    'gen_2/pokedex/',  
    show_progress=True,
    use_multithreading=True)

docs = loader.load()
len(docs)
```

```python
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=40)
documents = text_splitter.split_documents(docs, )
```

## Create vector database

```python
pip install openai
```

```python
db = Chroma.from_documents(documents, OpenAIEmbeddings(), persist_directory="chromadb")
```

```python
collection = db.get()
collection
```

```python
db.persist()
```

```python
query = "What is a good check for rest/talk raikou?"
raikou_docs = db.similarity_search(query)
raikou_docs
```

## RAG

```python
from langchain.chat_models import ChatOpenAI
from langchain.chains import RetrievalQA
```

```python
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.2)
```

```python
chain = RetrievalQA.from_chain_type(llm=llm,
                                    chain_type="stuff",
                                    retriever=db.as_retriever())

response = chain(query)
print(response['result'])
```

RAG output:


"A good check for RestTalk Raikou would be a Pokemon that can resist or take its attacks well and threaten it with super effective moves. Some potential checks for RestTalk Raikou could include Gengar, which can OHKO with Explosion, and Snorlax, which can 2HKO with Earthquake."
